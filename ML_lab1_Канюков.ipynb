```python
#Number of Doctors Visited
#The total count of different doctors the patient has seen = { 1: 0-1 doctors 2: 2-3 doctors 3: 4 or more doctors }
sns.kdeplot(npha['Number of Doctors Visited'],color='orange',fill=True)
plt.show()
sns.histplot(npha['Number of Doctors Visited'], bins=10, kde=True,color='green')
```

```python
#Age
#The patient's age group = { 1: 50-64 2: 65-80 }
sns.boxplot(y=npha['Age'])
plt.show()
sns.histplot(npha['Age'], bins=10,color='green')
```

```python
#Phyiscal Health
#A self-assessment of the patient's physical well-being = { -1: Refused 1: Excellent 2: Very Good 3: Good 4: Fair 5: Poor }
sns.kdeplot(npha['Phyiscal Health'],color='orange',fill=True)
plt.show()
sns.histplot(npha['Phyiscal Health'], bins=10, kde=True,color='green')
```
